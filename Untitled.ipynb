# 1. About Data
The dataset provides information on hotel reservations with 1,19,390 observations for a City Hotel and a Resort Hotel. Each observation represents a hotel booking between the 1st of July 2015 and 31st of August 2017, including booking that effectively arrived and booking that were canceled.

# 2. Business Problem and Assumptions
In recent years, both Resort Hotel and City Hotel have faced significant challenges due to high cancellation rates, leading to issues such as reduced revenues and suboptimal room utilization. The primary cause of suboptimal room usage is the elevated cancellation rates, resulting in decreased revenue generation. Lowering cancellation rates is crucial for both hotels to enhance efficiency and revenue.

As a data analyst's, our responsibility is to provide comprehensive business advice to address this pressing issue. The focus of this report is on analyzing hotel booking cancellations and identifying other factors unrelated to their business that may impact yearly revenue generation.

# 9. Analysis and Findings

**Now, first of all, it is important to see the percentage of cancellation over the whole dataset. We can do it with the help of values_counts() function from pandas. It is used to get a series containing counts of unique values in a DataFrame column.**

**When normalize is set to False (the default) , the result will contain the counts of unique values in the specified column.**
**When normalize is set to True , the result will contain the relative frequencies or percentages of each unique value in the specified column.**

**The percentages are calculated by dividing the count of each unique value by the total number of non-null values in the column.**

```python
cancellation_percentage = df['is_canceled'].value_counts(normalize = True)
```

```python
cancellation_percentage
```

**In the above results, 0 represents the not-canceled ratio, while 1 represents the cancellation ratio. Let's extract the percentages for 0 and 1 to print the results separately, and visualize in plots for clear understanding.**

```python
not_cancelled_percentage = cancellation_percentage[0] * 100
cancelled_percentage = cancellation_percentage[1] * 100
```

```python
print(f"No-Cancellation Ratio: {not_cancelled_percentage:.2f}%")
print(f"Cancellation Ratio: {cancelled_percentage:.2f}%")
```

>**9.1 No-Cancellation Ratio vs Cancellation**

```python
plt.rc('font', size=12)
plt.rc('axes', labelcolor='#000000')
plt.rc('text', color='#000000')
plt.rc('xtick', color='#000000')

categories = ['No Cancellation Ratio', 'Cancellation Ratio']
cancel_ratio = [not_cancelled_percentage, cancelled_percentage]
colors = ['#1069e8', '#e81123']
width = 0.5

ax = sns.barplot(x=categories, y=cancel_ratio, palette=colors, saturation=width)

plt.xlabel('Categories', labelpad=20)
plt.ylabel('Percentages', labelpad=20)
plt.title('Fig 1: No-Cancellation Ratio vs Cancellation Ratio', fontsize=20, pad=20)

ax.set_facecolor("#ededed")

plt.show()
```

>**Comparing Prices Trends of Cancellation and No-Cancellation**

```python
no_cancellation_data = df[df['is_canceled'] == 0]
```

```python
no_cancellation_data.head()
```

```python
no_cancellation_prices = no_cancellation_data.groupby('reservation_status_date')[['adr']].mean()
no_cancellation_prices.reset_index(inplace=True)
no_cancellation_prices.sort_values('reservation_status_date', inplace=True)
```

```python
no_cancellation_prices.head()
```

```python
no_cancellation_data = df[df['is_canceled'] == 0]

no_cancellation_data.head()

no_cancellation_prices = no_cancellation_data.groupby('reservation_status_date')[['adr']].mean()
no_cancellation_prices.reset_index(inplace=True)
no_cancellation_prices.sort_values('reservation_status_date', inplace=True)

no_cancellation_prices.head()
```

```python
cancellation_data = df[df['is_canceled'] == 1]
```

```python
cancellation_data.head()
```

```python
cancellation_prices = cancellation_data.groupby('reservation_status_date')[['adr']].mean()
cancellation_prices.reset_index(inplace=True)
cancellation_prices.sort_values('reservation_status_date', inplace=True)
```

```python
cancellation_prices.head()
```

```python
plt.rc('font', size=18)

plt.figure(figsize=(20, 8))

plt.gca().set_facecolor('#19a01a')

plt.plot(no_cancellation_prices['reservation_status_date'], 
         no_cancellation_prices['adr'], label='Non-Cancelled Prices', color='#f3ba00')

plt.plot(cancellation_prices['reservation_status_date'], 
         cancellation_prices['adr'], label='Cancelled Prices', color='#e81123')

plt.title('Fig 11: Average Daily Rate', fontsize=30, pad=30)
plt.xlabel('Timeperiod', labelpad=20)
plt.ylabel('Prices', labelpad=20)

plt.legend(fontsize=20)
plt.show()
```

**In the above wave graph, it's evident that the waves exhibit distinct patterns, and the data appears to be spread out. This spread is causing inconsistency in current data. To address this and enhance consistency, let's filter and utilize the data between September 2015 and September 2017.**

```python
cancellation_prices_filtered = cancellation_prices[
    (cancellation_prices['reservation_status_date']>'2015-09')
    & (cancellation_prices['reservation_status_date']<'2017-09')]

no_cancellation_prices_filtered = no_cancellation_prices[
    (no_cancellation_prices['reservation_status_date']>'2015-09')
    & (no_cancellation_prices['reservation_status_date']<'2017-09')]
```

**Now, after filtering the data, let's re-visualize the wave graph.**

```python
plt.rc('font', size=18)

plt.figure(figsize=(20, 8))

plt.gca().set_facecolor('#19a01a')

plt.plot(no_cancellation_prices_filtered['reservation_status_date'], 
         no_cancellation_prices_filtered['adr'], label='Non-Cancelled Prices', color='#f3ba00')

plt.plot(cancellation_prices_filtered['reservation_status_date'], 
         cancellation_prices_filtered['adr'], label='Cancelled Prices', color='#e81123')

plt.title('Fig 12: Average Daily Rate', fontsize=30, pad=30)
plt.xlabel('Timeperiod', labelpad=20)
plt.ylabel('Prices', labelpad=20)

plt.legend(fontsize=20)

plt.show()
```

**As observed, the prices of canceled reservation ratios are higher than the no-cancellation ratios, indicating that more cancellations occur when prices are elevated. This confirms that the highest prices are indeed impacting the reservation ratio.**

**1. To mitigate reservation cancellations, hotels can adjust pricing strategies by lowering rates based on specific locations. Offering discounts to consumers is another effective approach.**

**3. Initiating marketing campaigns in January, coupled with reasonable pricing, can help hotels increase revenue during a month with the highest cancellation rates.**
